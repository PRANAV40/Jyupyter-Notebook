```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
arr = np.array([1,2,3,4,5])
plt.plot(arr)
plt.show()
```

```python
arr = np.array([90, 180, 270, 360])

x = np.deg2rad(arr)
print(x)
plt.plot(x)
plt.show()
```

```python
arr = np.array([0,np.pi/4,np.pi/3,np.pi/2])
degree = np.degrees(arr)
sina = np.sin(arr)
plt.plot(degree,sina)
plt.show()
```
