```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
year = np.array([2014,2015,2016,2017,2018])
sales= np.array([2000,3000,4000,3500,6000])

plt.subplot(1,3,1)
plt.plot(year,sales,marker='o',color='Red',)
plt.title("Sales Income")
plt.xlabel('year')
plt.ylabel('Sales')

#plot 2
plt.subplot(2,3,3)
plt.plot(year,sales,marker='o',color='Orange')
plt.title("Purchase view")
plt.xlabel('year')
plt.ylabel('Purchase')
plt.show()
```

```python
year = np.array([2014,2015,2016,2017,2018])
sales= np.array([2000,3000,4000,3500,6000])
sizes = np.array([20,50,100,200,420])

plt.subplot(1,2,1)
plt.title("Sales Income")
plt.xlabel('year')
plt.ylabel('Sales')
plt.scatter(year,sales, s=sizes)
plt.show()
```
