```python
import numpy as np 
import matplotlib.pyplot as plt  
 
x = np.arange(0, 3 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

plt.plot(x, y) 
plt.show() 
```

```python
from matplotlib import pyplot as plt
  
# x-axis values
x = [5, 2, 9, 4, 7]
  
# Y-axis values
y = [10, 5, 8, 4, 2]
  
# Function to plot the bar
plt.bar(x,y)
  
# function to show the plot
plt.show()
```
