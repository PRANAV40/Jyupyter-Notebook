```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = [1, 7, 2]

myseries = pd.Series(a)

print(myseries)
```

```python
a = [1, 7, 2]

myseries = pd.Series(a, index = ["x", "y", "z"])

print(myseries)
```

```python
marks = {"12th": 420, "10th": 380, "Graduation": 390}

myseries = pd.Series(marks)

print(myseries)
```

```python
data = {
  "Sales": [420, 380, 390],
  "Profit": [50, 40, 45]
}

df = pd.DataFrame(data)

print(df)
```

```python
data = {
  "Sales": [420, 380, 390],
  "Profit": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(df) 
```

```python
data = [["April",1.47],["May",1.53],["June",1.45],["July",1.49],["Agust",1.62]]
df = pd.DataFrame(data,columns=["Months","Collection"])
print(df)
data1=df['Months']
data2=df['Collection']
plt.title("GST Collections per months")
plt.xlabel("Months")
plt.ylabel("Collections")

plt.plot(data1,data2,)
plt.show()
```

```python
data = [["Mandeep",20],["Rahul",70],["Sandeep",50],["Rohan",40]]
df = pd.DataFrame(data,columns=["Students","Marks"])
print(df)
data1=df['Students']
data2=df['Marks']
plt.title("Students Marks in a Subject")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.bar(data1,data2)
plt.show()
```
