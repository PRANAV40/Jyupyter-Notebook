## Histogram (hist) function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_bins = 10
a = np.random.randn(2000,3)
fig, ((ax0, ax1),(ax2, ax3)) = plt.subplots(nrows=2, ncols=2)

colors = ['red','blue','green']
ax0.hist(a, a_bins, density = True, histtype='bar', color=colors, label=colors)
ax0.legend(prop={'size':12})
ax0.set_title("BARS")

ax1.hist(a, a_bins, density=True, histtype="bar",stacked=True)
ax1.set_title('Stacked Bars')

ax2.hist(a, a_bins, histtype="step",stacked=True, fill=False)
ax2.set_title("Stack step")

a_multi = [np.random.randn(b) for b in [1000,5000,10000]]
ax3.hist(a_multi, a_bins, histtype="bar")
ax3.set_title("Diff Samples")

fig.tight_layout()
plt.show()
```
