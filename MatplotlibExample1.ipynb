```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Product = ['A','B','C','D','E']
Sales = [249,349,199,249,299]
Profit = [29,48,21,32,24]
a = np.arange(len(Product))
width = 0.34
fig, ax = plt.subplots()
a1 = ax.bar(a-width/2, Sales, width, label="Sales")
a2 = ax.bar(a+width/2, Profit, width, label="Profit")
ax.set_xlabel("Product")
ax.set_ylabel("Sales and Profit")
ax.set_title("Sales and Profit")
ax.set_xticks(a)
ax.set_xticklabels(Product)
ax.legend()
fig.tight_layout()
plt.show()
```

```python
Product = ['A','B','C','D','E']
Sales = [249,349,199,249,299]
Profit = [29,48,21,32,24]
a = np.arange(len(Product))
width = 0.34
fig, ax = plt.subplots()
a1 = ax.barh(a-width/2, Sales, width, label="Sales")
a2 = ax.barh(a+width/2, Profit, width, label="Profit")
ax.set_xlabel("Product")
ax.set_title("Sales and Profit")
ax.set_xticks(a)
ax.set_xticklabels(Product)
ax.legend()
fig.tight_layout()
plt.show()
```

```python
Product = ['A','B','C','D','E']
Sales = [249,349,199,249,299]
Profit = [29,48,21,32,24]
plt.subplot(2,1,1)
plt.bar(Product,Sales)
plt.title("Product with Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
plt.subplot(2,1,2)
plt.bar(Product,Profit)
plt.title("Product with Profit")
plt.xlabel("Product")
plt.ylabel("Profit")
plt.show()
```
