## Categorical Variable Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
data = {'Monday':15,'Tuesday':7,'Wednesday':20}
days = list(data.keys())
km_run = list(data.values())
fig, axs = plt.subplots(1,3,figsize=(10,5),sharey=True)
axs[0].scatter(days,km_run)
axs[1].plot(days,km_run)
axs[2].bar(days,km_run)
fig.suptitle('Categorical Plots')
```
